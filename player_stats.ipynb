```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt

def read_df(file):
    df = pd.read_csv(file, sep = ';', encoding = 'ISO-8859-1') 
    return df

def info_df(df):
    display(df.head())
    display(df["Pos"].info())
    display(df["Pos"].value_counts())
    
     
def clean_df(df):
    display(df[df["Pos"] == "GKMF"])
    #Luca Zidane is actually a goalkeeper
    df["Pos"] = df["Pos"].replace("GKMF", "GK")
    
    # I recognized most of the players as FW
    display(df[df["Pos"] == "FWDF"].head(10))
    df['Pos'] = df['Pos'].replace('FWDF', 'FW')
    
    #I recognized most of the players as DF
    display(df[df["Pos"] == "DFFW"].head(10))
    df['Pos'] = df['Pos'].replace('DFFW', 'DF')

    #I recognized most of the players as MF    
    display(df[df["Pos"] == "MFDF"].head(10))
    df['Pos'] = df['Pos'].replace('MFDF', 'MF')
    
    #I recognized most of the players as DF
    display(df[df["Pos"] == "DFMF"].head(10))
    df['Pos'] = df['Pos'].replace('DFMF', 'DF')
    
    #I recognized most of the players as MF        
    display(df[df["Pos"] == "MFFW"].head(10))
    df['Pos'] = df['Pos'].replace('MFFW', 'MF')

    
    # I recognized most of the players as FW    
    display(df[df["Pos"] == "FWMF"].head(10))
    df['Pos'] = df['Pos'].replace('FWMF', 'FW')
    
    return df

    
def remove_value(df):
    df.replace([np.inf, -np.inf], np.nan, inplace=True)
    df.dropna(inplace=True)
    
    return df
    
    
########################### Machine Learning section ###########################

def split_df(df):
    features = df.select_dtypes(include=['float', 'int'])
    features.drop(['Rk', 'Age', 'Born', 'MP', 'Starts', 'Min', '90s'], axis=1, inplace=True)
    x = np.array(features)
    y = np.array(df["Pos"])
    xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
    return xtrain, xtest, ytrain, ytest

def train_model(xtrain, ytrain):
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(xtrain, ytrain)
    return model
    
    
def evaluate_model(model, xtest, ytest):
    predictions = model.predict(xtest)

    accuracy = accuracy_score(ytest, predictions)
    print("\n","\n","\n","Accuracy:", accuracy)
    return predictions
    
    
def combine(ytest, predictions):
    combined = pd.concat([pd.Series(ytest), pd.Series(predictions)], axis=1)
    combined.columns = ['actual', 'predictions']
    display(combined)

    
    
    
########################### Data Visualization section ###########################

def change(df):
    df['Goals'] = (df['Goals'] * df['90s']).round(0).astype(int)
    display(df['Goals'])
    return df

def grouping(df):
    goals_per_comp = df.groupby('Comp')['Goals'].sum() 
    print(goals_per_comp)
    
    avg_age = df.groupby('Comp')['Age'].mean()
    print(avg_age)
    return goals_per_comp, avg_age
    
    

def bar(df, y, xlabel, ylabel, title):
    plt.bar(y.index, y, color ='maroon', width = 0.4)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    
    plt.show()



def radar_graph(df):
    players = ['Bruno Fernandes', 'Bernardo Silva','Kevin De Bruyne']

    selected_players_df = df[df['Player'].isin(players)]

    attrs = ['PasTotCmp%', 'PasShoCmp%', 'PasMedCmp%', 'PasLonCmp%', 'PasAss', 'PPA', 'PasProg']
    selected_players_df[attrs] = selected_players_df[attrs] / selected_players_df[attrs].max()

    num_attrs = len(attrs)

    plt.figure(figsize=(8, 8))
    angles = np.linspace(0, 2 * np.pi, num_attrs, endpoint=False)
    angles = np.concatenate((angles, [angles[0]])) 

    for i, player in enumerate(selected_players_df['Player']):
        values = selected_players_df.loc[selected_players_df['Player'] == player, attrs].values.tolist()[0]
        values += values[:1] 
        plt.polar(angles, values, marker='o', label=player)
        plt.fill(angles, values, alpha=0.3)
    
    plt.title('Skills Comparison for Players')
    plt.xticks(angles[:-1], attrs)
    plt.legend()
    plt.show()
 



    
def main():
###########################  Reaading in and cleaning the data ###########################
    
    file = "player_stats.csv"
    df = read_df(file)
    
    info_df(df)
    
    df = clean_df(df)
    
    df = remove_value(df)

    print("\n","\n","\n","Cleaned Dataset","\n","\n","\n")
    display(df["Pos"].value_counts())

###########################  Machine Learning model ###########################
       
    
    xtrain, xtest, ytrain, ytest = split_df(df)
    model = train_model(xtrain, ytrain)
    predictions = evaluate_model(model,xtest,ytest)
    
    combine(ytest, predictions)

    
    print('\n','\n','\n','\n')
    
    
###########################  Reaading in and cleaning the new data ###########################
    
    
    file1 = "player_stat_22_23.csv"
    df1 = read_df(file1)
    
    info_df(df1)
    
    df1 = clean_df(df1)
    
    df1 = remove_value(df1)
    
    print("\n","\n","\n","Cleaned Dataset","\n","\n","\n")
    display(df1["Pos"].value_counts())

    print('\n','\n','\n','\n')

    
###########################  Machine Learning model ###########################

    xtrain1, xtest1, ytrain1, ytest1 = split_df(df1)
    model1 = train_model(xtrain1, ytrain1)
    predictions1 = evaluate_model(model1,xtest1,ytest1)
    
    combine(ytest1, predictions1)
    





###########################  Visualizations ###########################

    df = change(df)
    
    print('\n','\n','21/22 Season Visualization','\n','\n')
    goals_per_comp, avg_age = grouping(df) 
    
    xlabel = 'Competition'
    y1label = 'Goals'
    title1 = 'Goals per Competition'
    bar(df, goals_per_comp, xlabel, y1label, title1)
    
    y2label = 'Age'
    title2 = 'Avg Age per Competiton'
    bar(df, avg_age, xlabel, y2label, title2)
    
    
    print('\n','\n','22/23 Season Visualization','\n','\n')

    goals_per_comp1, avg_age1 = grouping(df1) 
    
    xlabel = 'Competition'
    y1label = 'Goals'
    title1 = 'Goals per Competition'
    bar(df1, goals_per_comp1, xlabel, y1label, title1)
    
    y2label = 'Age'
    title2 = 'Avg Age per Competiton'
    bar(df1, avg_age1, xlabel, y2label, title2)
    

    radar_graph(df)
    

main()
```
